## Decision Trees

deals with understanding the working of decision trees.

```python
# Importing libraries in Python
import sklearn.datasets as datasets
import pandas as pd
import matplotlib.pyplot as plt
# Loading the iris dataset
iris=datasets.load_iris()
# Forming the iris features data
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head())
# Forming the target data
y=iris.target
print(y)
```

```python
#importing DecisionTree algorithm as dtree
from sklearn.tree import DecisionTreeClassifier 
dtree=DecisionTreeClassifier()
#Fitting the Decision Tree
dtree.fit(df,y)

# Visualizing the Tree

#Improting the tree to plot the Decision Trees
from sklearn import tree
# resize the figure to be easily recongized
plt.figure(figsize=(10,10))
# filling the tree 
tree.plot_tree(dtree, filled=True)
# Title the figure
plt.title("Decision trees")
plt.show()
```
